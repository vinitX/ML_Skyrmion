```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import time
import pandas as pd
```

```python
t=time.time()
s = np.array(pd.read_csv("3spin.csv",header=None)).astype("float32")
H = np.array(pd.read_csv("3H.csv",header=None)).astype("float32")
H2 = np.array(pd.read_csv("3H2.csv",header=None)).astype("float32")
s=np.reshape(s,(-1,100))
print(time.time()-t)
print(np.shape(H),np.shape(H2),np.shape(s))
```

```python
C=[]
C_=[]
Q=[]
L=100
length=500
k=(len(H)//length)
t=time.time()
for i in range(length):
    C.append((np.mean(H2[i*k:(i+1)*k])-np.mean(H[i*k:(i+1)*k])**2)/(2-2*i/length)**2)
    TP=(2-2*i/length)
    C_.append((1-np.tanh(1/TP)**2)/(TP**2))
    Q.append(np.mean(H[i*k:(i+1)*k]))
print(time.time()-t)
plt.plot(C)
plt.plot(np.array(C_)*4*L)
plt.show()
plt.plot(Q)
plt.show()
```

```python
corr=np.zeros((length,L,L))
t=time.time()
N=np.shape(s)[0]//length
for T in range(length):
    print(T,end=' ')
    sr=np.zeros((1,L))
    srr=np.zeros((L,L))
    
    for m in range(T*N,(T+1)*N):
        x=np.reshape(s[m],(1,L))
        sij=np.matmul(np.transpose(x),x)
        r=np.zeros((1,L))
        for n in range(-L+1,L):
            r[0,n%L]+=np.trace(sij,offset=n)
        sr=sr+r
        srr=srr+np.matmul(np.transpose(r),r)

    sr=sr/N
    srr=srr/N

    corr[T,:,:]=srr-np.matmul(np.transpose(sr),sr)
print(time.time()-t)
```

```python
lr=3

temp=np.arange(2,0,-2/length)
```

```python
import scipy.ndimage.filters as filt
corrf=np.zeros((length,lr,lr))

for i in range(lr):
    for j in range(lr):
        corrf[:,i,j]=filt.gaussian_filter(corr[:,i+1,j+1],20)
              
Y=np.zeros(length)

for i in range(length):
    Y[i]=C[i]*temp[i]**(2)

Yf=filt.gaussian_filter(Y,20)
```

```python
#Gradient Descent

hist=[]
d_hist=[]
h_hist=np.zeros((1000,lr))

#h=np.zeros(5)
h=(np.random.rand(lr)-0.5)*0.01

alpha=1e-4 #3e-5
t=time.time()
for lsteps in range(100):
    print(lsteps,end=' ')
    for i in range(lr):
        for step in range(1000):
            tem=np.sum(h*np.matmul(corrf,h),axis=1)-Yf

            f=np.sum(tem**2)/(length*length)
            tem=np.reshape(tem,(length,1))
            df=np.sum(4*tem*np.matmul(corrf,h),axis=0)/(length*length)
    
            h[i]=h[i]-alpha*df[i]
            if np.abs(f)>1e10: break
            hist.append(f)
            d_hist.append(alpha*df)
        h_hist[lsteps,:]=h

print(time.time()-t)
```

```python
plt.plot(hist)
plt.show()
plt.plot(d_hist[:3000])
plt.show()
plt.plot(d_hist)
plt.show()
plt.plot(h_hist[:100])
plt.show()
```
